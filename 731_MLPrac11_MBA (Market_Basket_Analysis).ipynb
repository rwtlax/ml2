```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = [
    ["Milk","Onion","Nutmeg","Kidney Beans","Eggs","Yogurt"],
    ["Chips","Onion","Nutmeg","Kidney Beans","Eggs","Yogurt"],
    ["Milk","Corn Flour","Sweet Corn""Kidney Beans","Yogurt"],
    ["Milk","Onion","Spring Onion","Kidney Beans","Ice Cream","Eggs"]
]
```

```python
data
```

```python
type(data)
```

```python
te = TransactionEncoder() # encoding
te_array = te.fit(data).transform(data)
te_array
```

```python
df = pd.DataFrame(te_array,columns=te.columns_) # chips liya hai toh true nahi toh false
df
```

```python
freq_item = apriori(df,min_support=0.6,use_colnames=True) # support calculate hoga
freq_item
```

```python
# Association Rule Mining
arm = association_rules(freq_item,metric="confidence",min_threshold=0.7)
arm
```

```python
rule = arm[['antecedents','consequents','support','confidence']] # to extract multiple columns [[]]
```
