# Aim :- Implementation of LASSO Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.datasets import make_regression
```

```python
x,y=make_regression(n_samples=100,n_features=5,n_informative=3,n_targets=1,noise=20,random_state=4)
```

```python
x
```

```python
y
```

```python
alphas = [0.001,0.01,0.1,1,2,3,4,5,6,15,20,30,50,80]
for i in alphas:
    l=Lasso(alpha=i)
    l.fit(x,y)
    print(l.coef_) # beta ki values     
```

```python
from sklearn.datasets import load_diabetes
cf=load_diabetes(as_frame=True)
p=cf.data
p
```

```python
q=cf.target
q
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(p,q,test_size=0.25,random_state=2)
```

```python
alphas = [0.001,0.01,0.1,1,2,3,4,5,6,15,20,30,50]
for i in alphas:
    l=Lasso(alpha=i)
    l.fit(x_train,y_train)
    print(l.coef_) # beta ki values 
```

**Conclusion** : When we increase the value of alpha the value of coefficient gets penalized ( becomes zero ) hence, it leads to automatic feature selection
