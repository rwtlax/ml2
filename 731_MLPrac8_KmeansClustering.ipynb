```python
import warnings 
warnings.filterwarnings("ignore")
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("student_clustering.csv")
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['ML'])
```

```python
# Find no. of Predefined clusters
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)    
```

```python
wcss
```

```python
# To find Elbow point
plt.plot(range(1,11),wcss) # Elbow point is 4 therefore 4 clusters 
```

```python
x = df.iloc[:,:].values  # centeroid
km = KMeans(n_clusters=4)
y_means = km.fit_predict(x)
```

```python
y_means
```

```python
x[y_means==0,0] # cgpa
```

```python
x[y_means==0] # cluster 0
```

```python
x[y_means==0,1] # ML
```

```python
plt.scatter(x[y_means==0,0],x[y_means==0,1],c='r')# plot for cluster 0
plt.scatter(x[y_means==1,0],x[y_means==1,1],c='b')# plot for cluster 1
plt.scatter(x[y_means==2,0],x[y_means==2,1],c='g')# plot for cluster 2
plt.scatter(x[y_means==3,0],x[y_means==3,1],c='y')# plot for cluster 3
```
