# Aim : - Implementation of Density Based Spacial Clustering of Applications with Noise (DBSCAN) Algorithm.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN # In practical exam hierarchical & k-means dono karo and then dbscan
```

```python
df = pd.read_csv("DBSCAN.csv")
df.head()
```

```python
plt.figure(figsize = (10,7))
plt.scatter(df['A'],df['B'])
```

```python
db = DBSCAN(eps = 30 , min_samples = 5) # When eps=0.5 it considers only one cluster for all datapoints
```

```python
db.fit(df[['A','B']])
```

```python
df['Label'] = db.labels_
```

```python
df['Label'].unique()
```

```python
plt.figure(figsize = (10,7))
plt.scatter(df['A'],df['B'],c = df['Label'])
```
