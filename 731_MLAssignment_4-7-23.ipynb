```python
median1=new_df1.median(numeric_only=True)
df5=new_df1.replace(0.00, median1)
```

```python
x = df[['Cost of Revenue',
 'Earnings Before Interest and Tax',
 'Earnings Before Tax',
 'Gross Profit',
 'Income Tax',
 'Net Income-Cont. Operations',
 'Sales, General and Admin.',
 'Total Current Liabilities'
 ]]
x
```

```python
y=df[['Total Revenue']]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import Lasso
```

```python
alphas = [0.001,0.01,0.1,1,2,3,4,5,6,15,20,30,50,100]
for i in alphas:
    l=Lasso(alpha=i)
    l.fit(x_train,y_train)
    print(l.coef_) # beta ki values 
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt

# Create and fit the Lasso regression model
lasso_model = Lasso(alpha=0.1)  # Adjust the alpha value
lasso_model.fit(x_train, y_train)

# Predict using the Lasso model
y_pred = lasso_model.predict(x_test)

# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print('R-squared:', r2)
```
